## 1.Sales Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sales_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Product_A': [150, 200, 170, 220, 210, 250],
    'Product_B': [120, 180, 160, 200, 190, 230],
    'Product_C': [100, 130, 120, 150, 140, 170],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South']
}

df = pd.DataFrame(sales_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot for Product A, B, and C sales across months
df_melted = df.melt(id_vars='Month', value_vars=['Product_A', 'Product_B', 'Product_C'], 
                    var_name='Product', value_name='Sales')

sns.barplot(x='Month', y='Sales', hue='Product', data=df_melted)
plt.xticks(rotation=45)
plt.title('Sales of Products Across Months')
plt.show()
```

```python
# Line plot for sales trends over months
sns.lineplot(x='Month', y='Product_A', data=df, label='Product A', marker='o')
sns.lineplot(x='Month', y='Product_B', data=df, label='Product B', marker='o')
sns.lineplot(x='Month', y='Product_C', data=df, label='Product C', marker='o')

plt.title('Sales Trends of Products Over Months')
plt.xticks(rotation=45)
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Boxplot for sales distribution of products
sns.boxplot(data=df[['Product_A', 'Product_B', 'Product_C']])
plt.title('Sales Distribution of Products')
plt.ylabel('Sales')
plt.show()
```

```python
# Correlation heatmap between products
correlation = df[['Product_A', 'Product_B', 'Product_C']].corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Product Sales')
plt.show()
```

## 2 .Employee Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame
employee_data = {
    'Employee_ID': [101, 102, 103, 104, 105, 106],
    'Department': ['HR', 'Sales', 'IT', 'Finance', 'Marketing', 'HR'],
    'Monthly_Score': [85, 90, 78, 88, 95, 92],
    'Attendance': [22, 23, 21, 24, 23, 25],  # out of 25 days
    'Projects_Completed': [3, 4, 2, 5, 4, 3]
}
df = pd.DataFrame(employee_data)

# Bar plot for average monthly score per department
sns.barplot(x='Department', y='Monthly_Score', data=df)
```

```python
# Scatter plot for Monthly Score vs Attendance
sns.scatterplot(x='Attendance', y='Monthly_Score', data=df, hue='Department', style='Department', s=100)
plt.title('Monthly Score vs Attendance (by Department)')
plt.xlabel('Attendance (days)')
plt.ylabel('Monthly Score')
plt.show()
```

```python
# Box plot for Monthly Score distribution by Department
sns.boxplot(x='Department', y='Monthly_Score', data=df)
plt.title('Monthly Score Distribution by Department')
plt.show()
```

```python
# Correlation heatmap
correlation = df[['Monthly_Score', 'Attendance', 'Projects_Completed']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Employee Metrics')
plt.show()
```

```python
# FacetGrid for Monthly Score distribution across Departments
g = sns.FacetGrid(df, col='Department', col_wrap=3)
g.map(sns.histplot, 'Monthly_Score')

g.add_legend()
plt.show()
```

## 3. Course Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame
course_data = {
    'Course_Name': ['Python for Beginners', 'Advanced Machine Learning', 'Data Visualization', 
                    'Web Development', 'Mobile App Development', 'AI and Robotics'],
    'Week_1_Rating': [4.5, 4.7, 4.3, 4.6, 4.1, 4.8],
    'Week_2_Rating': [4.6, 4.8, 4.4, 4.5, 4.2, 4.9],
    'Week_3_Rating': [4.7, 4.6, 4.5, 4.7, 4.3, 4.8],
    'Total_Reviews': [150, 230, 180, 210, 160, 300]
}
df = pd.DataFrame(course_data)

# Bar plot for total reviews
sns.barplot(x='Course_Name', y='Total_Reviews', data=df)
plt.title('Total Reviews for Each Course')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Prepare data for line plot by melting the DataFrame
df_melted = df.melt(id_vars='Course_Name', value_vars=['Week_1_Rating', 'Week_2_Rating', 'Week_3_Rating'], 
                    var_name='Week', value_name='Rating')

# Line plot for course ratings across weeks
sns.lineplot(x='Week', y='Rating', hue='Course_Name', data=df_melted, marker='o')
plt.title('Course Ratings Over 3 Weeks')
plt.show()
```

```python
# Prepare data for heatmap
rating_data = df.set_index('Course_Name')[['Week_1_Rating', 'Week_2_Rating', 'Week_3_Rating']]

# Create a heatmap for ratings
sns.heatmap(rating_data, annot=True, cmap='YlGnBu', cbar=True)
plt.title('Course Ratings Heatmap (Weeks 1-3)')
plt.show()
```

```python
# Box plot for course ratings over the weeks
sns.boxplot(x='Course_Name', y='Rating', data=df_melted)
plt.title('Distribution of Course Ratings over 3 Weeks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# FacetGrid for Rating distribution across Weeks
g = sns.FacetGrid(df_melted, col='Course_Name', col_wrap=3)
g.map(sns.histplot, 'Rating')

g.add_legend()
plt.show()
```
