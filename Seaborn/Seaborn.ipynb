## 1. Distribution Plots

### 1.1. `sns.histplot()` (Histogram)

- **Purpose**: Displays the distribution of a single variable using bars to represent the frequency of values.
- **Usage**: Useful for showing the frequency of different ranges (bins) of a continuous variable.


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Example with 'tips' dataset
tips = sns.load_dataset('tips')
sns.histplot(data=tips, x='total_bill', kde=True)  # kde=True adds a Kernel Density Estimate (smooth curve)
plt.show()
````

- **Customizations**:
  - `bins`: Number of bins (bar groups).
  - `hue`: Grouping variable for color encoding.
  - `kde`: Shows density estimation on top of the histogram.

### 1.2. `sns.kdeplot()` (Kernel Density Estimation)

- **Purpose**: Displays the distribution of a variable using a smooth continuous line. Good for seeing the probability distribution of data.

````python
sns.kdeplot(data=tips, x='total_bill', shade=True,fill=True)
plt.show()
````

- **Customizations**:
  - `shade`: Fills the area under the KDE curve.
  - `bw_adjust`: Adjusts the bandwidth (smoothing) of the KDE.

### 1.3. `sns.rugplot()` (Rug Plot)

- **Purpose**: Adds small vertical lines (rugs) along the x-axis to show individual observations.
- **Usage**: Used in combination with other distribution plots.

````python
sns.rugplot(data=tips, x='total_bill')
plt.show()
````

## 2. Categorical Plots

### 2.1. `sns.barplot()` (Bar Plot)

- **Purpose**: Shows the mean (average) of a numeric variable for different categories.
- **Usage**: Great for comparing the mean values of a numeric variable across categories.


````python
sns.barplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
````

- **Customizations**:
  - `hue`: Splits the data by an additional categorical variable.
  - `ci`: Confidence interval (set to None to remove).
  - `palette`: Set the color palette.

### 2.2. `sns.countplot()` (Count Plot)

- **Purpose**: Displays the count of observations in each categorical bin using bars.
- **Usage**: Good for showing the frequency distribution of a categorical variable.

````python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
````

- **Customizations**:
  - `hue`: Adds another layer of separation using a second categorical variable.
  - `order`: Change the order of categories shown.


### 2.3. `sns.boxplot()` (Box Plot)

- **Purpose**: Shows the distribution of a numeric variable across different categories, including outliers.
- **Usage**: Used to display the median, quartiles, and outliers of data.


````python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
````

- **Customizations**:
  - `hue`: Adds grouping by another categorical variable.
  - `palette`: Change the colors of the violin plot.
  - `width`: Change the width of the violins.

### 2.4. `sns.violinplot()` (Violin Plot)

- **Purpose**: Similar to a box plot but also shows the KDE of the distribution.
- **Usage**: Combines box plot and KDE to provide more detailed insights about the distribution.

````python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
plt.show()
````

- **Customizations**:
  - `split`: If True, shows the KDE for each hue category side-by-side in one violin.

### 2.5. `sns.stripplot()` (Strip Plot)

- **Purpose**: Displays all data points for a numeric variable in a categorical context as individual points.
- **Usage**: Useful for seeing every data point (not summarized like a bar plot).

````python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', jitter=True)
plt.show()
````

- **Customizations**:
  - `jitter`: Adds random noise to the data points so they dont overlap.
  - `hue`: Adds color grouping.

## 3. Relational Plots

### 3.1. `sns.scatterplot()` (Scatter Plot)

**Purpose:** Shows the relationship between two numeric variables.

**Usage:** Useful for identifying patterns, clusters, and correlations.


````python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
````

**Customizations:**

- `hue:` Color encoding for a categorical variable.
- `style:` Adds different markers for different categories.
- `size:` Adjust point size based on another numeric variable.

### 3.2. `sns.lineplot()` (Line Plot)

**Purpose:** Displays the relationship between two variables with a line connecting the data points.

**Usage:** Used for showing trends over time or other continuous variables.


````python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
````

**Customizations:**

- `markers:` Adds markers to each data point along the line.
- `hue, style, size:` Same customizations as in `sns.scatterplot()`.


## 4. Matrix Plots

### 4.1. `sns.heatmap()` (Heatmap)

**Purpose:** Displays a matrix of data values as a color-coded grid.

**Usage:** Useful for showing correlations or comparing variables over time.


````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a customized dataset
data = {
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] * 2,
    'year': [2021] * 12 + [2022] * 12,
    'value': [20, 30, 25, 35, 40, 45, 50, 55, 60, 65, 70, 75, 
              22, 32, 27, 37, 42, 47, 52, 57, 62, 67, 72, 77]
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Pivot the DataFrame
df_pivot = df.pivot(index='month', columns='year', values='value')

# Create a heatmap
sns.heatmap(df_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)

# Set titles and labels
plt.title('Customized Heatmap Example')
plt.xlabel('Year')
plt.ylabel('Month')

# Display the plot
plt.show()
````

**Customizations:**

- **cmap**: Set the color palette (e.g., 'Blues', 'coolwarm').
- **annot**: Annotates each cell with its value.
