## 1. Sales_Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating the DataFrame from the dictionary
sales_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Product_A': [150, 200, 170, 220, 210, 250],
    'Product_B': [120, 180, 160, 200, 190, 230],
    'Product_C': [100, 130, 120, 150, 140, 170],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South']
}



```

### Line Plot of Sales Over Months
A line plot can help visualize the sales trend over time for different products.

```python
df = pd.DataFrame(sales_data)


plt.figure(figsize=(8,6))
plt.plot(df['Month'], df['Product_A'], marker='o', label='Product A')
plt.plot(df['Month'], df['Product_B'], marker='o', label='Product B')
plt.plot(df['Month'], df['Product_C'], marker='o', label='Product C')
plt.title('Monthly Sales of Products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

### Bar Plot Comparing Sales of Products in Each Month
A bar plot can be used to compare the sales of different products in each month.

```python
# Set the plot size
plt.figure(figsize=(10,6))

# Plot the sales of products for each month using a bar chart
df.plot(x='Month', kind='bar', stacked=False ,)

# Add title and labels
plt.title('Sales Comparison of Products by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(title='Products')

plt.show()
```

### Heatmap for Sales Correlation Between Products
A heatmap is useful for visualizing the correlation between the sales of different products.

```python
# Calculate the correlation between the sales of different products
sales_corr = df[['Product_A', 'Product_B', 'Product_C']].corr()

# Plot the heatmap
plt.figure(figsize=(6,4))
sns.heatmap(sales_corr, annot=True, cmap='Blues')
plt.title('Correlation Between Product Sales')
plt.show()
```

### Scatter Plot of Product A vs Product B Sales
A scatter plot can help visualize the relationship between the sales of two products.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Product_A', y='Product_B', data=df, hue='Region', style='Region')
plt.title('Product A vs Product B Sales by Region')
plt.xlabel('Product A Sales')
plt.ylabel('Product B Sales')
plt.show()
```

### Bar Plot for Sales by Region
A bar plot can also be used to compare total sales by region.

```python
# Group the sales data by region and calculate total sales for each product
region_sales = df.groupby('Region').sum()

# Plot the sales by region using a bar plot
region_sales.plot(kind='bar', stacked=True, figsize=(8,6))
plt.title('Total Sales by Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Creating a pie chart
plt.pie(sizes, labels=labels)

# Adding a title
plt.title('Basic Pie Chart')
plt.show()
```

## 2.Employee_Data

```python
import pandas as pd 
import matplotlib.pyplot as plt

employee_data = {
    'Employee_ID': [101, 102, 103, 104, 105, 106],
    'Department': ['HR', 'Sales', 'IT', 'Finance', 'Marketing', 'HR'],
    'Monthly_Score': [85, 90, 78, 88, 95, 92],
    'Attendance': [22, 23, 21, 24, 23, 25],  # out of 25 days
    'Projects_Completed': [3, 4, 2, 5, 4, 3]
}


df = pd.DataFrame(employee_data)
```

```python
# Scatter plot between Monthly Score and Attendance
plt.scatter(df['Monthly_Score'], df['Attendance'], color='green')
plt.title('Monthly Score vs Attendance')
plt.xlabel('Monthly Score')
plt.ylabel('Attendance')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame(employee_data)

# Group by Department and find average Monthly Score
avg_score_by_dept = df.groupby('Department')['Monthly_Score'].mean()

# Plot bar chart
avg_score_by_dept.plot(kind='bar', color='skyblue')
plt.title('Average Monthly Score by Department')
plt.xlabel('Department')
plt.ylabel('Average Monthly Score')
plt.show()
```

```python
# Line plot to visualize attendance of each employee
plt.plot(df['Employee_ID'], df['Projects_Completed'], marker='o', linestyle='-', color='blue')
plt.plot(df['Employee_ID'], df['Monthly_Score'], marker='>', linestyle='--', color='g')
plt.plot(df['Employee_ID'], df['Attendance'], marker='*', linestyle='-.', color='r')
plt.title('Employee Attendance')
plt.xlabel('Employee ID')
plt.ylabel('Attendance (out of 25 days)')
plt.grid(True)
plt.show()
```

```python
# Stacked bar chart for Projects Completed and Attendance
plt.bar(df['Employee_ID'], df['Projects_Completed'], label='Projects Completed', color='blue')
plt.bar(df['Employee_ID'], df['Attendance'], bottom=df['Projects_Completed'], label='Attendance', color='orange')

plt.title('Projects Completed and Attendance by Employee')
plt.xlabel('Employee ID')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Box plot of Monthly Scores
plt.boxplot(df['Monthly_Score'], vert=False, patch_artist=True, boxprops=dict(facecolor='cyan'))
plt.title('Monthly Score Distribution')
plt.xlabel('Monthly Score')
plt.show()
```

```python
# Line plot to visualize attendance of each employee
plt.plot(df['Employee_ID'], df['Attendance'], marker='o', linestyle='-', color='blue')
plt.title('Employee Attendance')
plt.xlabel('Employee ID')
plt.ylabel('Attendance (out of 25 days)')
plt.grid(True)
plt.show()
```

## 3.Course Data

```python
course_data = {
    'Course_Name': ['Python for Beginners', 'Advanced Machine Learning', 'Data Visualization', 
                    'Web Development', 'Mobile App Development', 'AI and Robotics'],
    'Week_1_Rating': [4.5, 4.7, 4.3, 4.6, 4.1, 4.8],
    'Week_2_Rating': [4.6, 4.8, 4.4, 4.5, 4.2, 4.9],
    'Week_3_Rating': [4.7, 4.6, 4.5, 4.7, 4.3, 4.8],
    'Total_Reviews': [150, 230, 180, 210, 160, 300]
}
```

```python
df3 = pd.DataFrame(course_data)
df3.head(5)
```

```python

# Calculate average rating for each course
df3['Avg_Rating'] = df3[['Week_1_Rating', 'Week_2_Rating', 'Week_3_Rating']].mean(axis=1)

# Plot bar chart for average ratings
plt.bar(df3['Course_Name'], df3['Avg_Rating'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Average Course Ratings')
plt.xlabel('Course Name')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Line plot to show rating trends over the weeks for each course
plt.plot(df3['Course_Name'], df3['Week_1_Rating'], label='Week 1', marker='o')
plt.plot(df3['Course_Name'], df3['Week_2_Rating'], label='Week 2', marker='o')
plt.plot(df3['Course_Name'], df3['Week_3_Rating'], label='Week 3', marker='o')

plt.xticks(rotation=45, ha='right')
plt.title('Weekly Ratings for Each Course')
plt.xlabel('Course Name')
plt.ylabel('Rating')
plt.legend()
plt.show()
```

```python
# Scatter plot between Total Reviews and Average Rating
plt.scatter(df3['Total_Reviews'], df3['Avg_Rating'], color='green')
plt.title('Total Reviews vs Average Rating')
plt.xlabel('Total Reviews')
plt.ylabel('Average Rating')
plt.grid(True)

# Annotating the courses for clarity
for i, course in enumerate(df3['Course_Name']):
    plt.annotate(course, (df3['Total_Reviews'][i], df3['Avg_Rating'][i]))

plt.show()
```

```python
# Pie chart for the proportion of reviews by course
plt.pie(df3['Total_Reviews'], labels=df3['Course_Name'], autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'orange', 'pink', 'cyan', 'yellow'])
plt.title('Proportion of Total Reviews by Course')
plt.show()
```

```python
# Stacked bar chart for weekly ratings of each course
plt.bar(df3['Course_Name'], df3['Week_1_Rating'], label='Week 1', color='blue')
plt.bar(df3['Course_Name'], df3['Week_2_Rating'], bottom=df3['Week_1_Rating'], label='Week 2', color='green')
plt.bar(df3['Course_Name'], df3['Week_3_Rating'], bottom=df3['Week_1_Rating'] + df3['Week_2_Rating'], label='Week 3', color='orange')

plt.xticks(rotation=45, ha='right')
plt.title('Weekly Ratings by Course')
plt.xlabel('Course Name')
plt.ylabel('Rating')
plt.legend()
plt.show()
```

```python
# Box plot for weekly ratings distribution
plt.boxplot([df3['Week_1_Rating'], df3['Week_2_Rating'], df3['Week_3_Rating']], labels=['Week 1', 'Week 2', 'Week 3'], patch_artist=True)
plt.title('Weekly Ratings Distribution')
plt.ylabel('Ratings')
plt.show()
```
